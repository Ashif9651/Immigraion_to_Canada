```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('canadian_immegration_data.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.set_index('Country',inplace=True)
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
#list of years column
years = list(map(str, range(1980,2014)))
years
```

# visualizing data using matplotlib

```python
# In 2010, Haiti suffered a catastrophic magnitude 7.0 earthquake. The quake caused widespread devastation and loss of life and 
# aout three million people were affected by this natural disaster. As part of Canada's humanitarian effort, the Government of 
# Canada stepped up its effort in accepting refugees from Haiti. We can quickly visualize this effort using a `Line` plot:

# Question: Plot a line graph of immigration from Haiti using `df.plot()`.
```

```python
haiti = dataset.loc['Haiti',years]
haiti
```

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind='line')
plt.xlabel('years')
plt.ylabel('Number of immgrants')
plt.title('Immigration from Haiti')
plt.text(2000, 6000, '2010 Earthquake')
plt.show()
```

```python
# We can clearly notice how number of immigrants from Haiti spiked up from 2010 as Canada stepped up its efforts to 
# accept refugees from Haiti.
```

```python
#Observation 2 : comparing the number of immigrants from India and China from 1980 to 2013.
```

```python
df_IC = dataset.loc[['India','China'],years]
df_IC
```

```python
df_IC= df_IC.transpose()
df_IC
```

```python
df_IC.plot(kind='line')
plt.title('Immigrants from India and China from 1980 to 2013.')
plt.xlabel('years')
plt.ylabel('Number of Immigrants')
plt.show()
```

```python
# findings:  we can clearly see that form 1980 to around 1997 immigration of people from india was higher than china but form
# 1997 above china had higher immigration of people
```

```python
#Object 3 : Compare the trend of top 5 countries that contributed the most to immigration to Canada.
```

```python
top_5 = dataset.sort_values(['Total'], ascending=False, axis=0).head()
top_5
```

```python
top_5.drop(['Continent','Region','DevName','Total'],axis=1,inplace=True)
```

```python
top_5 = top_5.transpose()
top_5
```

```python

top_5.plot(kind='line',figsize=(14,8))
plt.title('top 5 countries that contributed the most to immigration to Canada')
plt.xlabel('years')
plt.ylabel('Number of Immigrants')
plt.show()
```

```python
#Object 3 : Compare the trend of least 5 countries that contributed the most to immigration to Canada.
```

```python
least_5 = dataset.sort_values(['Total'], ascending=False, axis=0).tail()
least_5
```

```python
least_5.drop(['Continent','Region','DevName','Total'],axis=1,inplace=True)
least_5
```

```python
least_5=least_5.transpose()
least_5
```

```python
least_5.plot(kind='line',figsize=(14,8))
plt.title('least 5 countries that contributed the most to immigration to Canada')
plt.xlabel('years')
plt.ylabel('Number of Immigrants')
plt.show()
```

```python
#compare the number of Icelandic immigrants (country = 'Iceland') to Canada from year 1980 to 2013.
df_I = dataset.loc['Iceland',years]
df_I.head()
```

```python
# Immigration to Canada by Continent [1980 - 2013]
```

```python
df_continents= dataset.groupby('Continent',axis=0).sum()
df_continents.head()
```

```python
df_continents['Total'].plot(kind='pie',
                           figsize=(5,6),
                           autopct='%1.1f%%',
                           shadow=True,
                            
                           
                           )    
plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal')
plt.show()
```

```python
 #ploting the box plot for the Japanese immigrants between 1980 - 2013.
```

```python
df_japan = dataset.loc[['Japan'], years].transpose()
df_japan.head()
```

```python
df_japan.plot(kind='box')
plt.title('Box plot of Japanese Immigrants from 1980 - 2013')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
df_CI_box = dataset.loc[['India','China'],years].transpose()
df_CI_box
```

```python
df_CI_box.plot(kind='box')
plt.title('Box plot of Japanese Immigrants from 1980 - 2013')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
fig = plt.figure()

ax0 = fig.add_subplot(1, 2, 1) 
ax1 = fig.add_subplot(1, 2, 2) 

# Subplot 1: Box plot
df_CI_box.plot(kind='box', color='blue', vert=False, figsize=(20, 6), ax=ax0) # add to subplot 1
ax0.set_title('Box Plots of Immigrants from China and India (1980 - 2013)')
ax0.set_xlabel('Number of Immigrants')
ax0.set_ylabel('Countries')

# Subplot 2: Line plot
df_CI_box.plot(kind='line', figsize=(20, 6), ax=ax1) # add to subplot 2
ax1.set_title ('Line Plots of Immigrants from China and India (1980 - 2013)')
ax1.set_ylabel('Number of Immigrants')
ax1.set_xlabel('Years')

plt.show()
```
